## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books, on="ISBN")
```

```python
ratings_with_name.head()
```

```python
num_ratings_df = ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_ratings_df.rename(columns={"Book-Rating": "Num_ratings"}, inplace=True)
num_ratings_df.head()
```

```python
avg_ratings_df = ratings_with_name.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_ratings_df.rename(columns={"Book-Rating":"Avg_Rating"},inplace=True)
avg_ratings_df.head()
```

```python
popularity_df = num_ratings_df.merge(avg_ratings_df, on="Book-Title")
popularity_df.head()
```

```python
popularity_df.describe()
```

```python
popular_df = popularity_df[popularity_df["Num_ratings"]>=250].sort_values("Avg_Rating",ascending=False).head(50)
popular_df.head()
```

```python
popular_df = popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M","Num_ratings","Avg_Rating"]]
popular_df.head()
```

## Collaborative Based Recommender System

```python
ratings_with_name
```

```python
#select users who have rated more than 200 books
x = ratings_with_name.groupby("User-ID").count()["Book-Rating"]>200
active_users = x[x].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name["User-ID"].isin(active_users)]
filtered_ratings.head()
```

```python
#select books which have more than 50 user ratings
y = filtered_ratings.groupby("Book-Title").count()["Book-Rating"]>=50
popular_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings["Book-Title"].isin(popular_books)]
final_ratings.head()
```

```python
pt = final_ratings.pivot_table(values="Book-Rating",columns="User-ID",index="Book-Title")
pt.fillna(0,inplace=True)
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
#function to recommend books

def recommend(book_name):
    
    #fetch index
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key= lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        
        item = []
        temp_df = books[books["Book-Title"] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        
        data.append(item)
        
    return data
```

```python
recommend("River's End")
```
